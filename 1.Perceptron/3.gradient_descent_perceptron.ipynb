### In this notebook, we will implement Simple Perceptron From Scratch.
- How the values of weights and bias is assigned and finally the best values are found using Gradient Descent Optimization technique.

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
# Let us build the dataset
X,y = make_classification(n_samples = 100 , n_features = 2 , n_informative= 1 ,
                        n_redundant = 0, n_classes= 2, n_clusters_per_class= 1, random_state = 12,
                        hypercube= False, class_sep= 14)
y = np.where(y==0,-1,1)
print(X.shape)
print(y.shape)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1] , c = y , cmap='viridis')
plt.show()
```

```python
def Perceptron(X,y,lr = 0.1,epochs = 1000):
    # initial values
    w1,w2,b = 1,1,1
    for e in range(epochs):
        for i in range(X.shape[0]):
            z = w1*X[i][0] + w2*X[i][1] + b
            if z*y[i] < 0:
                w1 = w1 + lr*(y[i] * X[i][0]) # this updates the 1st weight until minima is hit
                w2 = w2 + lr*(y[i] * X[i][1])
                b = b + lr*(y[i])
    return w1,w2,b
        
```

```python
w1,w2,b = Perceptron(X,y)
```

```python
print(w1)
print(w2)
print(b)
```

```python
# calculating slope and bias of the line
m = -(w1/w2)
c = -(b/w2)

print(f'Slope = {m}')
print(f'y-intercpet = {c}')
```

```python
# plot the line
plt.figure(figsize=(10,6))
x_line = np.linspace(-3,3,100)
y_line = m*x_line + c
plt.scatter(X[:,0],X[:,1] , c = y , cmap='viridis')
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.plot(x_line,y_line , color = 'red' , linestyle='--')
plt.show()
```
